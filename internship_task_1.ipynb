```python
# is prime function

def is_prime(n):
  if n in [2,3]:
    return True

  if n % 2 == 0 or n == 1 :
    return False

  r = 3
  while r * r <= n:
    if n % r == 0:
      return False
    r += 2
    return True
```

```python
is_prime(79)
```

```python
# Use of continue, break, pass in loop.

l = ['a', 'b', 'c', 'd']
for i in l:
  if i == 'a':
    continue
  elif i == 'b' or i == 'c':
    pass
  else:
    break
```

```python
# Example of function with and without return.

# factorial without recursive

def fact(n):
  prod = 1
  while n > 1:
    x = n
    n = n - 1
    prod = prod * x
  print(prod)

```

```python
fact(5)
```

```python
# factorial with recursive

def fact(n):
  if n == 1:
    return 1
  else:
    return n * fact(n-1)
```

```python
fact(5)
```

```python
from datetime import date

today = date.today()
print("Today's date:", today)

specific_date = date(2025, 5, 21)
print("Specific date:", specific_date)

timestamp = 1653120000 
date_from_timestamp = date.fromtimestamp(timestamp)
print("Date from timestamp:", date_from_timestamp)

print("Year:", today.year)
print("Month:", today.month)
print("Day:", today.day)

new_date = today.replace(year=2026, month=6, day=1)
print("New date:", new_date)

formatted_date = today.strftime("%d/%m/%Y")
print("Formatted date:", formatted_date)
```

```python
# Make following pattern using python.

# diamond
n = 3
for i in range(n):
    print(' ' * (n - i - 1) + '*' * (2 * i + 1))
for i in range(n - 2, -1, -1):
    print(' ' * (n - i - 1) + '*' * (2 * i + 1))
```

```python
# right shifted triangle with only odd values.

for i in range(1, 10, 2):
    for j in range(1, i + 1,2):
        print(j, end=' ')
    print()
```
